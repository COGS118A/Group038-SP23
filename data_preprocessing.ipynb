```python
import pandas as pd
from sklearn.utils import shuffle
import numpy as np
```

```python
# Get original data
df = pd.read_csv('data/diabetes_prediction_dataset.csv')
df = df.dropna(how='any')
df.head()
```

```python
# Transform categorical columns to numerical columns with one-hot-encoding
gender_one_hot = pd.get_dummies(df['gender'])
smoking_one_hot = pd.get_dummies(df['smoking_history'])
gender_one_hot = gender_one_hot.apply(lambda x: x.apply(int))
smoking_one_hot = smoking_one_hot.apply(lambda x: x.apply(int))

df = pd.merge(left=df, right=gender_one_hot, left_index=True, right_index=True)
df = pd.merge(left=df, right=smoking_one_hot, left_index=True, right_index=True)
df = df.drop(columns=['gender', 'smoking_history'])
df.head()
```

```python
# Shuffle the data and save it to a new csv file
np.random.seed(38)
df = shuffle(df)
df = df.reset_index(drop=True)
df.head()
```

```python
# Save the data to a separate csv file
df.to_csv('data/preprocessed_data.csv')
```
